```python
#practical implementation
#import required libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
#univariate anlysis
#we shall take one feature-petallength
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_virginica=df.loc[df['species']=='virginica']
```

```python
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.xlabel('Petal length')
plt.show()
#plotting using matplotlib functions
#blue is for setosa
#green is for virginica 
#ornage is for versi color
```

```python
#bivariate analysis
#here we take two features 
#in the above analysis you took only one feature
```

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

```python
#here we can easily classify setosa 
#but there is little ambiguity in classifying versicolor and virginica
#hence we must use multivariate analyysis
```

```python
#multivariate analysis
#pair plot in seaborn is useful
sns.pairplot(df,hue="species",size=1)
#if the plot is not clear you canuse size to modify
```
